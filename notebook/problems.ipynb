### Exercise 1

We will use Numpy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import statistics as stats
#1-
elementos_n =np.random.normal(size=100)
print(f'Elementos con distribución normal:', elementos_n)
#2
elementos_ch = np.random.chisquare(3,size=100)
print(f'Elementos con distribución chi-square:', elementos_ch)
#3.1 Calcular Medidas de tendencia central a-MEDIA/ b-MEDIANA/c-MODA
#a-
Media_n= elementos_n.mean()
print(f'Media de elementos con distribución normal : ', Media_n)
Media_ch = elementos_ch.mean()
print(f'Media de elementos con distribución chi-square : ', Media_ch)

#b-
Mediana_n = stats.median(elementos_n)
print(f'Mediana de elementos con distribución normal : ', Mediana_n)
Mediana_ch = stats.median(elementos_ch)
print(f'Mediana de elementos con distribución chi-square : ', Mediana_ch)

#c-
Moda_n = stats.mode(elementos_n)
print(f'Moda de elementos con distribución normal : ', Moda_n)
Moda_ch = stats.mode(elementos_ch)
print(f'Moda de elementos con distribución chi-square : ', Moda_ch)



#3.2- Calcular Medidas de dispersión a-RANGO/ b-VARIANZA c-/DESVIACIÓN ESTANDAR
#a-
Rango_n = max(elementos_n) - min(elementos_n)
print(f'Rango de elementos con distribución normal : ', Rango_n)
Rango_ch = max(elementos_ch) - min(elementos_ch)
print(f'Rango de elementos con distribución chi-square : ', Rango_ch)
#b-
Varianza_n = stats.variance(elementos_n)
print(f'Varianza de elementos con distribución normal : ', Varianza_n)
Varianza_ch = stats.variance(elementos_ch)
print(f'Varianza de elementos con distribución chi-square : ', Varianza_ch)
#c-
Desv_est_n = stats.stdev(elementos_n)
print(f'Desvío Estandar de elementos con distribución normal : ', Desv_est_n)
Desv_est_ch = stats.stdev(elementos_ch)
print(f'Desvío Estandar de elementos con distribución chi-square : ', Desv_est_ch)




#3.3- Calcular Medidas de forma a-SKEWNESS/ b-KURTOSIS
from scipy.stats import skew
#a-
Skew_n = skew(elementos_n)
print(f'Skew Normal:', Skew_n)
Skew_ch = skew (elementos_ch)
print(f'Skew Chisquare: ', Skew_ch)

#b-
from scipy.stats import kurtosis
Kurtosis_n = kurtosis(elementos_n)
print(f'Kurtosis Distribución NORMAL: ', Kurtosis_n)
Kurtosis_ch =kurtosis(elementos_ch)
print(f'Kurtosis Distribución CHISQUARE :', Kurtosis_ch)

````

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO

import math
import sys
#Definir la función de desvio estandar
# Paso 1: calcular la media.
# Paso 2: calcular el cuadrado de la distancia a la media para cada dato.
# Paso 3: sumar los valores que resultaron del paso 2.
# Paso 4: dividir entre el número de datos.
data = [4, 2, 5, 8, 6]

def desviciación_estandar(data,media):
    suma = 0
    for valor in data:
        suma += (valor-media)**2
    
    radicando = suma / (len(data)-1)

    return math.sqrt(radicando)






def calcular_media(data):
    suma = 0
    
    for valor  in data:
        suma += valor
   
    return suma / len(data)


media = calcular_media(data)
resultado = desviciación_estandar(data,media)
data = [4, 2, 5, 8, 6]
print(f"Muestra: {data}")
print(f"MEDIA: {media}")
print(f"Desviación Estandar: {resultado}")

    






````
